###### DataFrame Object :

Is a two dimensional table of data with column and row indexes (somthing like a spread sheet).

######  A DataFrame has Two indexes:
* The column index (df.columns): a list of string (variable names) or integers.
* The row_index (df.index): might be:
    * Integers - for case or row numbers;
    * String - for case names; or
    * DatetimeIndex or PeriodIndex - for time series.

## Missing_data
Replacing missing datas from the dataset with an appropriate value

```python
#importing nessesary libraries 
from sklearn.preprocessing import Imputer
```

```python
#assigning an object for Imputer
imputer = Imputer(missing_values='NaN', strategy='mean', axis=0) 
'''
for the strategy attribute -  median and most_frequent can also be used
axis 0 means it takes mean of the collumn, axis 1 will mean mean of the raw
'''
```

```python
df.head()
```

```python
dfarray = df.iloc[:, :].values
imputer = imputer.fit(dfarray[:, 3:7])
dfarray[:, 3:7] = imputer.transform(dfarray[:, 3:7])
t= pd.DataFrame(dfarray)
t.head()
```
